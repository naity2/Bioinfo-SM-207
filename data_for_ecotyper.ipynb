```python
import pandas as pd

from pathlib import Path
```

```python
# ZUMA-7 RNA-seq TPM

zuma7_tpm = pd.read_csv(
    Path.home() / "data/zuma7_signatures_log2/rnaseq_screening_tpm.csv",
    index_col=0
    )
zuma7_tpm.head()
```

##### Dataformat:
https://github.com/digitalcytometry/ecotyper  
The expression matrix should be in the TPM or FPKM space for bulk RNA-seq and non-logarithmic (exponential) space for microarrays. It should have gene symbols on the first column and gene counts for each sample on the next columns. Column (sample) names should be unique. Also, we recommend that the column names do not contain special characters that are modified by the R function make.names, e.g. having digits at the beginning of the name or containing characters such as space, tab or -. 

```python
zuma7_tpm = zuma7_tpm.T.reset_index()
zuma7_tpm.head()
```

```python
zuma7_tpm.rename(columns={"index": "GENES"}, inplace=True)

# replace "-"
zuma7_tpm.rename(columns=lambda c: c.replace("-", "_"), inplace=True)

zuma7_tpm.head()
```

```python
# check duplicate gene names
zuma7_tpm["GENES"].duplicated().any()
```

```python
# save
data_path = Path.home()/"data/bioinfo_sm207"
zuma7_tpm.to_csv(data_path/"zuma7_tpm.txt", sep="\t", index=False)
```

```python
# RCHOP
rchop_meta = pd.read_csv(
    Path.home() / "data/zuma7_signatures_log2/rchop_meta.csv",
    index_col=0
    )
rchop_data_tpm = pd.read_csv(
    Path.home() / "data/zuma7_signatures_log2/rchop_data_tpm.csv",
    index_col=0
    )
rchop_data_tpm.head()
```

```python
rchop_meta.head()
```

```python
schmitz_index = rchop_meta.loc[rchop_meta["Actual Arm Code"]=="Schmitz"].index
reddy_index = rchop_meta.loc[rchop_meta["Actual Arm Code"]=="Reddy"].index
```

```python
schmitz_tpm = rchop_data_tpm.loc[schmitz_index]
reddy_tpm = rchop_data_tpm.loc[reddy_index]
schmitz_tpm.shape, reddy_tpm.shape
```

```python
schmitz_tpm = schmitz_tpm.T.reset_index()
schmitz_tpm.rename(columns={"index": "GENES"}, inplace=True)

reddy_tpm = reddy_tpm.T.reset_index()
reddy_tpm.rename(columns={"index": "GENES"}, inplace=True)
```

```python
schmitz_tpm.head()
```

```python
reddy_tpm.head()
```

```python
# save
schmitz_tpm.to_csv(data_path/"schmitz_tpm.txt", sep="\t", index=False)
reddy_tpm.to_csv(data_path/"reddy_tpm.txt", sep="\t", index=False)
```
